В 14 столбцах есть пропуски, причем больше всего пропусков в столбцах is_apartment(20924), parks_nearest(15620), ponds_nearest (14589). Меньше всего пропущено в столбцах floors_total(86), locality_name(49). Проанализируем возможность и необходимость заполнения пропущенных значений в разных столбцах

Меньше всего пропусков в столбцах с названием населенного пункта (49 пропусков) и общим количеством этажей (86 пропусков). Однако заполнить пропущенные значения в этих столбцах не представляется возможным. Поскольку общее число пропущенных значений не достигает и 0,5%, а название населенного пункта и число этажей в доме - это важные параметры, представляется возможным удалить строки, содержащие пропуски в этих столбцах, чтобы в дальнейшем они не мешали нашему анализу.

```python
#перезаписываем датафрейм, удалив строки с пропусками в столбцах 'locality_name', 'floors_total'
data = data.dropna(subset=['locality_name', 'floors_total'])
```

**1.2. Заполним пропущенные значения там, где это возможно** \
\
В столбце 'is_apartment' 20805 пропусков. Логично предположить, что все эти пропуски означают, что квартира не является апартаментами. Поэтому заполним все эти пропуски значением False.

```python
#заполняем пропущенные значения в столбце 'is_apartment' на значение False
data['is_apartment'] = data['is_apartment'].fillna(False)
```

Заменим также пропущенные значения в столбце 'balcony'. Скорее всего столбец не заполняют в том случае, когда балкона нет. Поэтому заменим null на 0 

```python
#заполняем пропущенные значения в столбце 'balcony' на значение 0
data['balcony'] = data['balcony'].fillna(0)
```

Пустые значения можно заменить еще в двух столбцах 'parks_around3000', 'ponds_around3000'. Логично предположить, что в объявлении не указано количество парков и прудов в радиусе 3 км, поскольку их нет. Поэтому заменим пустые значения на 0.

```python
#заполняем пропущенные значения в столбцах 'parks_around3000', 'ponds_around3000' на значение 0
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
```

Пропуски в столбце, отвечающем за высоту потолков 'ceiling_height' можно заменить на медианное значение по всему датасету

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

Для того чтобы заполнить пропуски в столбцах 'kitchen_area' и 'living_area' можно вывести коэффициенты отношения медианного значения площади кухни к медианному значению общей площади и медианного значения жилой площади к медианному значению общей площади соответственно. И на основании этих коэффициентов по общей площади заполнить пропущенные значения.
 

```python
#рассчитываем коэффициент отношения медианной жилой площади к медианной общей площади
living_area_koef = data['living_area'].median()/data['total_area'].median()
#рассчитываем коэффициент отношения медианной кухонной площади к медианной общей площади
kitchen_area_koef = data['kitchen_area'].median()/data['total_area'].median()
# заменяем пустые значения в столбце 'living_area' на произведение общей площади и рассчитанного коэффициента
data['living_area'] = data['living_area'].fillna(data['total_area']*living_area_koef)
# заменяем пустые значения в столбце 'kitchen_area' на произведение общей площади и рассчитанного коэффициента
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area']*kitchen_area_koef)
```

Видим здесь выбросы, которые характеризуют недвижимость дороже 10 млн.
Для того чтобы рассчитать размах "усов" реализуем функцию, которая будет рассчитывать нижние и верхние границы "усов" для конкретного столбца


```python
def qvartile (col):
    # рассчитываем третий и первый квартили
    q3, q1 = np.percentile (data[col], [75 ,25])
    # рассчитываем межквартильный размах
    iqr = q3 - q1
    # считаем нижнюю границу
    low = q1-1.5*iqr
    # считаем верхнюю границу
    upper = q3+1.5*iqr
    return low,upper
```

Рассчитаем границы для всех выделенных нами столбцов

```python
# Помещаем в список столбцы с аномальными значениями
columns_anomaly = ['last_price', 'total_area', 'rooms', 'kitchen_area', 'living_area', 'ceiling_height', 'floors_total']
# В цикле для каждого столбца выводим границы
for col in columns_anomaly:
    print (f'Для столбца {col} нижняя и верхняя граница составит {qvartile(col)}')
```
